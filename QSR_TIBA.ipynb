```python
import pandas as pd
import numpy as np
df = pd.read_excel('QSR_DS.xlsx') 
df.head()
```

```python
# Check for missing values
print(df.isnull().sum())

# Percentage of missing values
missing_percentage = df.isnull().mean() * 100
print(missing_percentage)
```

```python
df.dropna(inplace= True)
df.info()
```

```python
print(f"Number of duplicate rows: {df.duplicated().sum()}")
```

```python
df['2023_RANK'] = df['2023_RANK'].str.rstrip('=')
df['2024_RANK'] = df['2024_RANK'].str.lstrip('=')
df['2023_RANK'] = df['2023_RANK'].str.rstrip('+')
df['2024_RANK'] = df['2024_RANK'].str.lstrip('+')
df
```

```python
print(df.columns)
```

```python
df['2023_RANK'] = pd.to_numeric(df['2023_RANK'], errors='coerce')
df['2024_RANK'] = pd.to_numeric(df['2024_RANK'], errors='coerce')
```

```python
def parse_rank_to_midpoint(rank):
    if '-' in str(rank):  # Check if rank is a range
        lower, upper = map(int, rank.split('-'))  # Extract lower and upper bounds
        return (lower + upper) // 2  # Calculate midpoint
    try:
        return int(rank)  # Convert single rank to integer
    except ValueError:
        return np.nan  # Handle unexpected values

df['2023_RANK'] = df['2023_RANK'].apply(parse_rank_to_midpoint)
df['2024_RANK'] = df['2024_RANK'].apply(parse_rank_to_midpoint)
```

```python
# Keep original ranks in new columns
df['2023_RANK_ORIGINAL'] = df['2023_RANK']
df['2024_RANK_ORIGINAL'] = df['2024_RANK']

# Apply midpoint transformation to the main rank columns
df['2023_RANK'] = df['2023_RANK'].apply(parse_rank_to_midpoint)
df['2024_RANK'] = df['2024_RANK'].apply(parse_rank_to_midpoint)
```

```python
# Flag rows where ranks were ranges
df['Is_Rank_Range'] = df['2023_RANK_ORIGINAL'].astype(str).str.contains('-') | df['2024_RANK_ORIGINAL'].astype(str).str.contains('-')
```

```python
df.dropna(inplace= True)
df.info()
```
